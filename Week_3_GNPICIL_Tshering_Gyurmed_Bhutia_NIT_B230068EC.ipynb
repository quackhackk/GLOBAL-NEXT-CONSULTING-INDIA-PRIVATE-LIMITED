```python
import pandas as pd
import numpy as np
df = pd.read_csv("spotify_dataset.csv")
```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
df = df.dropna()
```

```python
drop_cols = [
    'Unnamed: 0',
    'track_id',
    'artists',
    'album_name',
    'track_name'
]
df = df.drop(columns=drop_cols)
```

```python
df.head()
```

```python
print(df['track_genre'].nunique())
df['track_genre'].value_counts().head(20)
```

```python
features = [
    'danceability',
    'energy',
    'loudness',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo'
]
import matplotlib.pyplot as plt

for col in features:
    plt.figure(figsize=(5,3))
    df[col].hist(bins=30)
    plt.title(col)
    plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))

sns.heatmap(
    df[features].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA()

pca.fit(X_scaled)

print(
    np.cumsum(
        pca.explained_variance_ratio_
    )
)
```

```python
from sklearn.cluster import KMeans
inertia = []
for k in range(2,11):
    km = KMeans(
        n_clusters=k,
        random_state=42
    )
    km.fit(X_scaled)
    inertia.append(km.inertia_)
```

```python
plt.plot(
    range(2,11),
    inertia,
    marker='o'
)
plt.xlabel("K")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sample_df = df.sample(10000, random_state=42)
X_sample = scaler.fit_transform(
    sample_df[features]
)
for k in range(2,11):
    km = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    labels = km.fit_predict(X_sample)
    score = silhouette_score(
        X_sample,
        labels
    )
    print(f"K={k}, Score={score:.3f}")
```

```python
kmeans = KMeans(
    n_clusters=5,
    random_state=42,
    n_init=10
)

labels = kmeans.fit_predict(X_scaled)

df['cluster'] = labels
```

```python
cluster_profile = (
    df.groupby('cluster')[features]
      .mean()
)

print(cluster_profile)
```

```python
sample_df = df.sample(5000, random_state=42)

X_sample = sample_df[features]

scaler = StandardScaler()

X_sample_scaled = scaler.fit_transform(X_sample)
```

```python
from sklearn.cluster import SpectralClustering
spectral = SpectralClustering(
    n_clusters=5,
    affinity='nearest_neighbors',
    random_state=42
)
spectral_labels = spectral.fit_predict(
    X_sample_scaled
)
```

```python
for k in range(2,11):
    spectral = SpectralClustering(
        n_clusters=k,
        affinity='nearest_neighbors',
        random_state=42
    )
    labels = spectral.fit_predict(X_sample_scaled)
    score = silhouette_score(
        X_sample_scaled,
        labels
    )
    print(f"K={k}, Score={score:.3f}")
```
